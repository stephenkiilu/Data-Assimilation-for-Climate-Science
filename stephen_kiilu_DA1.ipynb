```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.optimize as sp
from scipy.optimize import fsolve
```

```python
def function(x):
    return 32*x**3-48*x**2+17.08*x-0.54
   
```

```python
  x=np.linspace(0,1,100)
    
```

```python
y=function(x)
root = fsolve(function,[0,1,2])
root
```

```python
plt.plot(x,y)
plt.axhline(y=0.0, color="black", linestyle=":")
plt.grid()
plt.title('Plot of the process model')
plt.xlabel('X_2')
```

```python
def fun(x2):
    y2=0.5
    y3=0.01
    return (y2-x2)**2+(y3-4*x2+4*x2**2)**2
```

```python
  x2=np.linspace(-0.15,1.15,50)
```

```python
yy= fun(x2)
root = fsolve(fun,[0,1,2])
root
```

```python
plt.plot(x2,yy)
plt.axhline(y=0.22, color="black", linestyle=":")
plt.grid()
plt.title('Plot of the cost function')
plt.xlabel('X_2')
```
