```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def bayes(mu,sigma_x,sigma_y,n,H,y):
    K=sigma_x@H.transpose()@np.linalg.inv(sigma_y+H@sigma_x@H.transpose())
    mu_star=mu+K@(y-H@mu)
    sigma_star=(np.identity(n)-K@H)@sigma_x
    return  sigma_star ,mu_star
    
    
```

```python
n=12
sigma_x=0.1*np.identity(n)
sigma_y=0.25*np.identity(n)
mu=np.array([76.9,91,114.2,154.2,88.1,18.6,11.4,31.1,69.6,105.7,112.7,77.4])
y=np.array([80,90,110,160,80,20,10,40,70,100,110,80])
H=np.identity(12)
sigma_star ,mu_star= bayes(mu,sigma_x,sigma_y,n,H,y)

```

```python
mu_star
```

```python
t=np.linspace(1,12,12)
plt.plot(t,mu)
plt.plot(t,y)
plt.plot(t,mu_star)
plt.show()
```
