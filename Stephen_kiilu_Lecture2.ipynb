```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
X=np.zeros(4)
X[0]=0.2
for j in range (0,3):
    X[j+1]=4*X[j]*(1-X[j])
print(X)
```

```python
#Er=np.zeros(2)
mean=0
sd=0.01
N=2
Er=np.random.normal(mean,sd,N)
```

```python
Y=X[1:3]+Er[1:3]
Y
```

```python
plt.plot(Y,"b*",markersize=10)
```

```python
x2=np.linspace(-0.1,1.1,100)
J=(x2-Y[0])**2+(4*x2*(1-x2)-Y[1])**2
J
plt.plot(x2,J)
```
