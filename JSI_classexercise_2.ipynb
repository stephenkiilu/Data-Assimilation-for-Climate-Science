```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
X = np.zeros(4)
X[0]= 0.2
for j in range(0,3):
    X[j+1] = 4*X[j]*(1-X[j])
print (X)
```

```python
np.random.seed(123)
me = 0
si = 0.001
N = 2
error = np.random.normal(me,si, N)
y = X[1:3] + error[1:3]
e_new =  np.random.normal(me,0.9, N)
y_new =  X[1:3] + e_new[1:3]
print(y,y_new)
```

```python
plt.plot(y,"b*" ,markersize=20)
plt.show()
```

```python
## Alternative question 2
me = 0
si = 0.001
N = 4
error = np.random.normal(me,si, N)
X = np.zeros(4)
Y  = np.zeros(4)
X[0]= 0.2
for j in range(0,3):
    X[j+1] = 4*X[j]*(1-X[j])
    Y[j] = X[j] + error[j]

plt.plot(Y[1:3],  "b*", markersize=10)
```
